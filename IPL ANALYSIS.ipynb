```python
import pandas as pd

matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")

print(matches.head())
print(deliveries.head())
```

```python
print(matches.shape)
print(matches.info())

print(deliveries.shape)
print(deliveries.info())
```

```python
matches.drop_duplicates(inplace=True)

matches['winner'].fillna('No Result', inplace=True)
```

```python
matches['winner'] = matches['winner'].fillna('No Result')
```

```python
print(matches['winner'].isnull().sum())
```

```python
print("Matches duplicates:", matches.duplicated().sum())
print("Deliveries duplicates:", deliveries.duplicated().sum())
```

```python
print(matches.columns)
print(deliveries.columns)
```

```python
team_wins = matches['winner'].value_counts()
print(team_wins.head(10))
```

```python
top_batsmen = deliveries.groupby('batter')['batsman_runs'].sum()
print(top_batsmen.sort_values(ascending=False).head(10))
```

```python
top_bowlers = deliveries.groupby('bowler')['is_wicket'].sum()
print(top_bowlers.sort_values(ascending=False).head(10))
```
